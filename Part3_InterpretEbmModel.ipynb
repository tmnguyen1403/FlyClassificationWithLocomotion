```python
import pandas as pd
import pickle
from matplotlib import pyplot as plt
```

```python
def plotAverage(scores,index,window_size,plot_title):
    t = pd.DataFrame(scores, index=index)
    rolling = t.rolling(window=window_size)
    rolling_mean = rolling.mean()
    t.plot(title=plot_title)
    #t.title(plot_title)
    average_title = f"Average with window size {window_size} {plot_title}"
    rolling_mean.plot(color='red', title=average_title)
    #rolling_mean.title(average_title)
    plt.show()
```

```python
def loadFeatureScore(file):
    data = None
    with open(file,"rb") as pickle_file:
        data = pickle.load(pickle_file)
    return data
```

```python
def handleScore(file):
    data = loadFeatureScore(file)
    for key in data.keys():
        index, scores = data[key]
        plotTitle = f'{key} {file.replace(".json","")}'
        plotAverage(scores,index,5,plotTitle)
```
