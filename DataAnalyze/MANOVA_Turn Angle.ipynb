```python
import pandas as pd
from statsmodels.multivariate.manova import MANOVA
```

```python
df = pd.read_excel("Combined Data in Turn Angle with Time Blocks.xlsx")
```

```python
df.columns = df.columns.str.replace(" ", "_")
```

```python
df
```

```python
maov = MANOVA.from_formula('Period * Fly_Type ~ Turn_Angle', data=df)
```

```python
print(maov.mv_test())
```

We can conclude that there is an association between turn angle and the relationship between period and genotype

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm
```

```python
model_lin = sm.OLS.from_formula("Turn_Angle ~ Period * Fly_Type", data=df)
result_lin = model_lin.fit()
result_lin.summary()
```

We reject the null hypothesis in this analysis is that a Turn Angle has no effect on Period and Fly Type at alpha level .05 because the p-values are all less than .05
